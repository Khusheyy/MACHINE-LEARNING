```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()  # first 5 rows
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()  # first 5 rows
```

```python
flights = sns.load_dataset('flights')
flights.head()  
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

1> scatterplot [numerical-numerical]

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
# shows that there is a linear rel bw total bill and tip
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex']) 
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker']) 
```

2 > bar plot [numerical-categorical]

```python
sns.barplot(x=titanic['pclass'], y=titanic['age'])
```

```python
sns.barplot(x=titanic['pclass'], y=titanic['fare'], hue=titanic['sex'])
```

```python
sns.barplot(x=titanic['pclass'], y=titanic['fare'], hue=titanic['sex'])
```

3 > box plot [numerical-categorical]

```python
sns.boxplot(x=titanic['sex'], y=titanic['age'],hue=titanic['survived'])
```

4 > distplot [numerical-categorical]

```python
sns.displot(titanic['age'], color='purple', kde=True)
```

```python
sns.displot(titanic[titanic['survived']==0]['age'], kind='kde')
sns.displot(titanic[titanic['survived']==1]['age'], kind='kde')
```

```python
sns.displot(data=titanic,x='age', hue='survived',kind='kde', fill=True, palette='Set1', alpha=.3)
plt.title('Age Distribution by Survival Status')
```

5 > heatmap [categorical-categorical]

```python
titanic.head()
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
# pclass and survived ka relation
```

```python
titanic['pclass'] = pd.to_numeric(titanic['pclass'], errors='coerce')
```

```python
titanic.groupby('pclass')['survived'].mean()*100
# 
```

```python
(titanic.groupby('pclass')['survived'].mean()*100).plot(kind = 'bar')
```

```python
titanic.groupby('sex')['survived'].mean()*100
```

```python
titanic.groupby('embarked')['survived'].mean()*100
```

6 > clustermap [categorical-categorical]

```python
sns.clustermap(pd.crosstab(titanic['pclass'],titanic['survived']))
```
